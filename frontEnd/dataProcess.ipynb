```python
house_price_dict = dict()
```

```python
for index, row in house_price.iterrows():
    if 'Melbourne' in row[' sa42016name']:
        # print(row[' dateyear'] + ' ' + row[' sa42016name'] + ': ' + row[' sold_both_auction_private_treaty_averageprice'])
        if row[' dateyear'] in house_price_dict:
            house_price_dict[row[' dateyear']] += float(row[' sold_both_auction_private_treaty_averageprice'])
        else:
            house_price_dict[row[' dateyear']] = float(row[' sold_both_auction_private_treaty_averageprice'])

        
        
```
